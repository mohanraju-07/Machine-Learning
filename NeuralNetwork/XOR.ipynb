```python
import pandas as pd
import numpy as np
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def sigmoid_derivative(z):
    return z*(1-z)

x =np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y =np.array([[0],[1],[1],[0]])

w1 =np.random.rand(2,2)
b1 = np.zeros((1,2))

w2 = np.random.rand(2,1)
b2 = np.zeros((1,1))

lr =0.1 

for epochs in range(50000):
    
    #forward 
    z1 = np.dot(x,w1) + b1 
    a1 = sigmoid(z1)

    z2 =np.dot(a1,w2) + b2
    a2 = sigmoid(z2)

    #loss 
    error = a2-y
    loss = np.mean(error**2)

    #back propagation 
    #output 
    dz2 = error * sigmoid_derivative(a2) 
    dw2 = np.dot(a1.T,dz2) / len(x)
    db2 = np.sum(dz2,axis=0, keepdims=True)/len(x)

    #hidden 
    dz1 = np.dot(dz2, w2.T) * sigmoid_derivative(a1)
    dw1 = np.dot(x.T,dz1) /len(x)
    db1 = np.sum(dz1,axis=0, keepdims=True)/len(x)

    # update 
    w2 -= lr * dw2
    b2 -= lr * db2
    w1 -= lr * dw1
    b1 -= lr * db1 

    if epochs % 5000 == 0:
        print("epoch:", epochs, "loss:", loss)
 
 # Final prediction
print("\nPredictions:")
print(a2)


```
