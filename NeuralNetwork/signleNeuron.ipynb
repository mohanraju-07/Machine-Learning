```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
#y=2x+1 
x =np.array([2,1])
y =np.array([5])

w =np.array([0.5,0.3])
b =1 
lr =0.01

z =np.dot(x,w) +b
loss =np.mean((y-z)**2)
print(loss)
while(loss>0.1):
    z =np.dot(x,w) +b
    error =y-z

    dl_dw =-2 * error * x
    dl_db = -2 *error 

    w = w -lr * dl_dw
    b = b - lr *dl_db

    print("weight: ",w)
    print("bais: ",b)

    loss = np.mean(error**2)
    print("loss: ",loss)

print("Final weights:", w)
print("Final bias:", b)


#value=sigmoid(z)
#print("value after activation: ",value)
```
