```python
import numpy as np

x = np.array([1,2,3,4])
y = np.array([5,7,9,11])

w = 0.01
b = 0.01
lr = 0.1

for epoch in range(1000):
    z = w * x + b         # prediction (no activation)
    
    error = z - y

    dw = np.mean(error * x)
    db = np.mean(error)

    w -= lr * dw
    b -= lr * db

    if epoch % 10 == 0:
        
        print(f"epoch {epoch} | loss={np.mean(error**2):.4f} | w={w:.4f} b={b:.4f}")
```

```python
import numpy as np

# Data: y = 2x + 1 + noise
X = np.array([[1], [2], [3], [4]])
y = np.array([[3], [5], [7], [9]])

# Initialize
w = 0.0  # weight
b = 0.0  # bias
lr = 0.01

# Train loop
for epoch in range(1000):
    # Forward
    z = np.dot(X, w) + b
    y_pred = z  # Linear output
    
    # Loss (MSE)
    loss = np.mean((y - y_pred)**2)
    
    # Backward
    dz = (y_pred - y) / len(y)
    dw = np.dot(X.T, dz) / len(y)
    db = np.sum(dz) / len(y)
    
    # Update
    w -= lr * dw
    b -= lr * db
    
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}, w: {float(w):.4f}, b: {float(b):.4f}")
```
